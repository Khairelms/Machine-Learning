```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs    # to provide dummy dataset look like clusters
```

```python
# Create a sample dataset using make_blobs

x,y = make_blobs(n_samples=300,       # it will create 300 fake data points .. x--> data point, y -->cluster
                 centers=4,           # no of cluster = 4
                 random_state=42)     # Use the same random behavior every time this code runs
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y)    # c=y --> assign diff diff colors based on Y
plt.title('Clusters Generated Using Make_Blobs')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()
plt.show()
```

```python
# Elbow Method
# let's assume we don't know the no. of cluster
# or I want to know no. of clusters (4) define above is correct or wrong

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    # kmeans now will apply algorithm steps on Dataset
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

print(wcss)

# client has given you a dataset

# X = [1,2,4,5]
# create 2 clusters
#      c1 = [1,2]
# 	 c2 = [4,5]
#   k = 2 correct K ?
#   elbow method
#     computes WCSS   WITHIN CLUSTER SUM OF SQUARED
# 	          SKLEARN (inertia_)
# ----------------------------------------------------///////


# compute the mean of c1   ==> 1.5

# compute the mean of c2   ==> 4.5


# compute squared distances of centroid

# (1 - 1.5)2  --> 0.25
# (2 - 1.5)2  --> 0.25
# (4 - 4.5)2  --> 0.25
# (5 - 4.5)2  --> 0.25

# in order to calculate inertia_
# 0.25 + 0.25 + 0.25 + 0.25   ==> 1.0

# INERTIA IS JUST A SCORE THAT TELL HOW TIGHTLY POINTS ARE GROUPED TOGETHER
#  SMALLER NUMBER = POINTS ARE CLOSER TO EACH OTHER
```

```python
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    # kmeans now will apply algorithm steps on Dataset
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)    # used ELBOW METHOD

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters (k)')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
# Predict New Point
new_point = [[ -9.29768866,   6.47367855]]
predict_cluster = kmeans.predict(new_point)
print('Predict_cluster : ', predict_cluster)
```
