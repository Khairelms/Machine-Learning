```python
import pyfpgrowth

# Step 1: Create transactions / import dataset
# 1st need covert to list of lists --> transactions = [[item for item in row if pd.notnull(item)] for row in df.values.tolist()]

transactions = [
 ['Milk', 'Bread', 'Butter'],
 ['Bread', 'Butter'],
 ['Milk', 'Egg'],
 ['Bread', 'Egg'],
 ['Bread', 'Butter', 'Egg', 'Milk']
]

print(transactions)
```

```python
# Step 2: FP-GROWTH FREQUENT PATTERNS

frequent_patterns = pyfpgrowth.find_frequent_patterns(transactions,support_threshold=2)
print("Frequent Patterns")
print(frequent_patterns)
```

```python
# Step 3: Generate Rules

rules = pyfpgrowth.generate_association_rules(frequent_patterns, confidence_threshold=0.5)
print("Association Rules")
print(rules)
```
